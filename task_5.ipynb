```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)  
y = 3 * x + 2          

plt.plot(x, y, marker='o', color='red', linestyle='-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart of y = 3x + 2")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["A", "B", "C", "D", "E"]
values = [15, 25, 30, 20, 10]

plt.bar(categories, values, color='Orange')

for i, val in enumerate(values):
    plt.text(i, val + 0.5, str(val), ha='center', fontsize=10)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart with Annotations")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100) 
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(x/4) 

plt.plot(x, y1, color='blue', linestyle='-', label='sin(x)')
plt.plot(x, y2, color='red', linestyle='--', label='cos(x)')
plt.plot(x, y3, color='black', linestyle=':', label='exp(x/4)')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines in One Plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])

fig, axs = plt.subplots(2, 2, figsize=(8, 6))

axs[0, 0].plot(x, y, marker='o', color='blue')
axs[0, 0].set_title("Line Plot")

axs[0, 1].bar(x, y, color='green')
axs[0, 1].set_title("Bar Chart")

axs[1, 0].scatter(x, y, color='red', s=80)
axs[1, 0].set_title("Scatter Plot")

data = np.random.randn(100)
axs[1, 1].hist(data, bins=10, color='orange', edgecolor='black')
axs[1, 1].set_title("Histogram")

fig.suptitle("2x2 Grid of Different Plots", fontsize=14)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)

product_A = np.array([5, 7, 8, 6, 7, 9, 10, 12, 11, 9, 10, 12])
product_B = np.array([3, 4, 4, 5, 6, 6, 7, 8, 7, 6, 8, 9])
product_C = np.array([2, 3, 3, 4, 4, 5, 5, 6, 5, 4, 6, 7])

plt.stackplot(months, product_A, product_B, product_C, 
labels=["Product A", "Product B", "Product C"], 
colors=["orange", "black", "green"])

plt.xlabel("Month")
plt.ylabel("Sales (in units)")
plt.title("Cumulative Sales of Products Over 12 Months")
plt.legend(loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Math": np.random.randint(50, 100, 20),
    "Science": np.random.randint(45, 95, 20),
    "English": np.random.randint(55, 98, 20)
       }

df = pd.DataFrame(data)
df.plot(kind='box', color={'boxes':'red', 'whiskers':'black', 'medians':'green', 'caps':'black'})
plt.title("Student Exam Scores Distribution")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

companies = ["Company A", "Company B", "Company C", "Company D"]
market_share = [35, 25, 20, 20]

explode_values = [0.1 if share == max(market_share) else 0 for share in market_share]

plt.pie(market_share, labels=companies, autopct="%1.1f%%", 
explode=explode_values, colors=["blue", "red", "green", "purple"], 
shadow=True, startangle=90)

plt.title("Market Share of Companies")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin(x)", color='blue')
plt.plot(x, y2, label="cos(x)", color='red')

plt.gca().invert_xaxis()

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Line Plot with Axis and Legend Customization")
plt.tight_layout()
plt.show()
```
